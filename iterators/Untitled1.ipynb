```python
import matplotlib.pyplot as plt
import csv

x=[]
y=[]

with open('test.txt','r') as csvfile:
    plots=csv.reader(csvfile)
    for column in plots:
        x.append(column[0])
        y.append(column[1])
plt.plot(x,y,label='File',color='red')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Test Graph")
plt.legend()
plt.show()
```
